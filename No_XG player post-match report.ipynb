```python
# ADD PLAYERS FORMATION
# READ FORMATIONS
formation = pd.read_excel(r"D:/Projects/football analysis/team foramion id (version 1).xlsx")

# ADD TEAMS FORMATIONS COLUMNS 
final_events['Team formation id']=final_events['Team formation id'].fillna(0)
final_events['Team formation id']=final_events['Team formation id'].astype('int64')
final_events['Involved']=final_events['Involved'].fillna('0')
final_events['Team player formation']=final_events['Team player formation'].fillna('0')

final_events['Team formation'] = pd.merge(final_events, formation, on='Team formation id', how='left')['formation'] 
final_events['Positions'] = pd.merge(final_events, formation, on='Team formation id', how='left')['positions']





# CREATE A TEMP DF TO HOLD TEAMS FORMATIONS AND ALL PLAYERS POSITIONS (combined_team_df)

# Initialize lists to store the dictionaries
home_team_data = []
away_team_data = []

# Filter the DataFrame to get the rows with 'Team set up'
team_setup_events = final_events[final_events['type/displayName'] == 'Team set up']

# Iterate through the filtered DataFrame
for i in range(0, len(team_setup_events), 2):
    # Home team dictionary
    home_team_dict = {
        "Involved": team_setup_events.iloc[i]["Involved"].split(', '),
        "Team player formation": team_setup_events.iloc[i]["Team player formation"].split(', '),
        "Positions": team_setup_events.iloc[i]["Positions"].split(', '),
        "Team formation": team_setup_events.iloc[i]["Team formation"],
        "Team id": team_setup_events.iloc[i]["teamId"]
    }
    home_team_data.append(home_team_dict)

    # Check if there is a corresponding away team row
    if i + 1 < len(team_setup_events):
        # Away team dictionary
        away_team_dict = {
            "Involved": team_setup_events.iloc[i + 1]["Involved"].split(', '),
            "Team player formation": team_setup_events.iloc[i + 1]["Team player formation"].split(', '),
            "Positions": team_setup_events.iloc[i + 1]["Positions"].split(', '),
            "Team formation": team_setup_events.iloc[i+1]["Team formation"],
            "Team id": team_setup_events.iloc[i+1]["teamId"]
        }
        away_team_data.append(away_team_dict)




def extract_positions(position_str):
    return position_str.split(',')

# Function to create player info dictionaries
def create_player_info(team_data):
    player_info = []
    for team in team_data:
        positions = extract_positions(team["Positions"][0])
        for involved, formation, position in zip(team["Involved"], team["Team player formation"], positions):
            player_info.append({
                "playerId": involved,
                "Team formation": team["Team formation"] ,
                "Team id":team['Team id'],
                "Team player formation": formation,
                "Position": position
                
            })
    return player_info

# Create player info for home team
home_team_players = create_player_info(home_team_data)
away_team_players = create_player_info(away_team_data)


# Convert list of dictionaries to DataFrame
home_team_df = pd.DataFrame(home_team_players)
away_team_df = pd.DataFrame(away_team_players)

# Display the DataFrame
# print("Home Team DataFrame")
# print(home_team_df)

# print("Away Team DataFrame")
# print(away_team_df)


combined_team_df = pd.concat([home_team_df, away_team_df], ignore_index=True)


# CREATE TEMP DF TO HOLD POSITIONS OF SUBBED ON PLAYERS (player_on_data)
player_on_data = final_events[final_events['type/displayName'] == 'Player on']
player_on_data = player_on_data.loc[:, [ 'playerId','teamId','Formation slot']]

player_on_data.rename(columns = {'Formation slot':'Team player formation','teamId':'Team id'}, inplace = True)

player_on_data = player_on_data.merge(combined_team_df[["Team player formation",'Team id', "Position"]],on = ["Team player formation", "Team id"], how ='left' )

player_on_data.rename(columns = {'Position':'New Position','Team player formation':'New formation'}, inplace = True)



# ADD SUBBED ON PLAYER POSITIONS TO ALL PLAYERS POSITIONS
combined_team_df =pd.merge(combined_team_df, player_on_data,  on=['playerId','Team id'], how='left')

combined_team_df["Team player formation"] = combined_team_df["New formation"].fillna(combined_team_df["Team player formation"])
combined_team_df["Position"] = combined_team_df["New Position"].fillna(combined_team_df["Position"])


combined_team_df.drop(columns=["New formation"], inplace=True)
combined_team_df.drop(columns=["New Position"], inplace=True)
combined_team_df.drop(columns=["Team id"], inplace=True)
combined_team_df.rename(columns = {'Team player formation':'Player formation'}, inplace = True)


final_events.drop(columns=["Positions","Team formation"], inplace=True)


# MERGE PLAYER POSITIONS TO FINAL_EVENTS
final_events= pd.merge(final_events, combined_team_df, on='playerId', how='left')
```
